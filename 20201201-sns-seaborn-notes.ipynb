```python
import seaborn as sns
dataset = sns.load_dataset('attention', index_col=0)
## barplot
sns.barplot(x=dataset['attention'], y=dataset['score'])
```

```python
sns.barplot(x='attention', y='score', data=dataset)
```

```python
sns.barplot(x='attention', y='score', hue='solutions',data=dataset)
```

```python
## pointplot
sns.pointplot(x='attention', y='score', hue='solutions', data=dataset)
```

```python
## distplot
dataset=sns.load_dataset('car_crashes')
sns.distplot(dataset['alcohol'])
```

```python
## jointplot
sns.jointplot('alcohol', 'no_previous', data=dataset)
```

```python
sns.jointplot('alcohol', 'no_previous', data=dataset, kind='reg', joint_kws={'order':2})
```

```python
# pairplot
dataset=sns.load_dataset('iris')
sns.pairplot(dataset)
```

```python
sns.pairplot(dataset, hue='species', vars=['sepal_width', 'sepal_length'])
```

```python
## boxplot
dataset=sns.load_dataset('tips')
sns.boxplot(x='day', y='tip', data=dataset, hue='smoker')
```

```python
## stripplot, swarmplot, violinplot, lvplot
```

```python
## countplot
dataset=sns.load_dataset('tips')
sns.countplot(x='day', data=dataset)
#sns.countplot(y='day', data=dataset, hue='smoker')
```

```python
## heatmap
dataset=sns.load_dataset('flights')
heatmap_data=dataset.pivot('year', 'month', 'passengers')
sns.heatmap(heatmap_data)
```

```python
sns.heatmap(heatmap_data, linewidth=1.0, annot=True, fmt='d')
```
